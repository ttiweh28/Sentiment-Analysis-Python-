```python
import wordcloud 
```

```python
list = []
fh = open('Bobi_tw.csv', 'r',encoding = 'utf8')

for line in fh:
    list.append(line.strip().split(','))    
   
```

```python
list
```

```python
from wordcloud import WordCloud 
import matplotlib.pyplot as plt

#convert list to string and generate
unique_string=''.join(map(str, list))
wordcloud = WordCloud(width = 1000, height = 500, background_color='white',).generate(unique_string)
plt.figure(figsize=(15,8))
plt.imshow(wordcloud)
plt.axis("off")
plt.savefig("wordcloud_twitter"+".png", bbox_inches='tight')
plt.show()
plt.close()
```
