```python
#Display Sub articles found ------ WORKING WELL ON 17TH AUGUST 2016 SEMI-AUTOMATIC
import urllib2
from urllib2 import urlopen
from bs4 import BeautifulSoup
import re
import csv
import sys

sourceFile=input("ENTER INPUT FILE FOR URLS")
f = open(sourceFile)
urls = csv.reader(f)
myList=[]
#print len(urls[0])
for row in urls:
    print row[0]
    myList.append(row[0])
print myList
print "&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&&"
print ""
print ""
print ""
filename = input ('ENTER OUT PUT FILE NAME:')
csvFile = open(filename, 'ab')
csvWriter = csv.writer(csvFile)

x=['DATES','TITLE ARTICLE','ARTICLE CONTENT']
csvWriter.writerow(x)
y=0
for ur in myList:
    url=ur
    source=urllib2.urlopen(url).read()

    articles=re.findall(r'<h3>(.*?)</h3>',source)
    for item in articles:
        #print '*****************************************************************'
        #print "ARTICLE TITLE"
        head=re.sub("<.*?>", "",item)
        #print head
        link=re.findall(r'href="(.*?)"',item)
        link="http://www.newvision.co.ug"+link[0]
        linkhtml=urllib2.urlopen(link).read()
        souplink = BeautifulSoup(linkhtml)
        #title=souplink.title.string
        #title=str(title)
        print '*********************************************************************************'
        print 'title:',head
        print '------------------------------'
        div = souplink.find('div', class_="publish-date")
        div=str(div)
        date = re.sub("<.*?>", "",div)
        date = re.sub('\s', ' ', date)
        print 'Date:', date
        print '------------------'
        #<div class="article-content"
        content=souplink.find('div', class_="article-content")
        content=str(content)
        content=textfinal1=re.sub("<.*?>", "",content)
        content=re.sub("\n","", textfinal1)
        y=y+1
        print content
        x=[date, head, content]
        csvWriter.writerow(x) 
        #output = re.sub('\s', ' ', text)
print 'NUMBER OF ARTICLES RETRIEVED:',y
```
