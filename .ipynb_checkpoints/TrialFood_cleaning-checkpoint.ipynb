```python
# Read csv dataset based on food key word from NewVision 2016
import pandas as pd
food_not_clean=pd.read_csv('2014HR1-14ALL.csv')
food_not_clean
```

```python
#Display the dimension- rows and columns of data
print food_not_clean.shape
```

```python
#Display the last 20 records and the first 20 record
food_not_clean.tail(20)
```

```python
#Inspect for missing values
food_not_clean.isnull().sum()
```

```python
#Remove record with missing values and the shape
food_not_clean_NoMissing=food_not_clean.dropna(how='any')
print food_not_clean_NoMissing.shape
```

```python
#Display first 3 records and identify any noise (uncessary data)
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][0]
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][1]
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][4]
```

```python
# Remove unwanted coding linked with googletag.cm.push
food_not_clean_NoMissing['ARTICLE CONTENT']=food_not_clean_NoMissing['ARTICLE CONTENT'].str.replace("(googletag\.cmd\.push\(.*?\);)"," ")
```

```python
#Display first 2 records and identify uncessary data AGAIN
food_not_clean_NoMissing['ARTICLE CONTENT'].head(2)
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][1]
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][4]
```

```python
#REMOVE UNCESSARY SPACES AND NEWLINES
food_not_clean_NoMissing['ARTICLE CONTENT']=food_not_clean_NoMissing['ARTICLE CONTENT'].str.replace("\s+", " ")
```

```python
#Display content without execess spaces and newlines 
food_not_clean_NoMissing['ARTICLE CONTENT'].head(2)
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][1]
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][4]
```

```python
#REMOVE UNCESSARY DESCENTANTS OF googletag.cm.push
food_not_clean_NoMissing['ARTICLE CONTENT']=food_not_clean_NoMissing['ARTICLE CONTENT'].str.replace("\-\-\>.*?\-\-\>", " ")
food_not_clean_NoMissing['ARTICLE CONTENT'].head(2)
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][1]
print '***************************************************'
print food_not_clean_NoMissing['ARTICLE CONTENT'][4]
print food_not_clean_NoMissing['ARTICLE CONTENT'].head(10)
```

```python
# Display the shape of the dataset
food_not_clean_NoMissing.shape
```

```python
#Inspect data for any missing data
food_not_clean_NoMissing.isnull().sum()
```

```python
#SAVE A CLEANED FILE TO a csv file
food_not_clean_NoMissing.to_csv('Hunger_2014_cleaned.csv')
```
